### Question 2c [time estimate: 10 minutes]

Write a Python implementation of the MAX-HEAPIFY operation using the pseudocode above, and your newly written functions, `left` and `right`.

```python
def heapify(heap, i):
    """
    Parameters
    ----------
    heap : list of floats
        Assume that the heap size is the length of the heap
        
    Note
    ----
    No output is needed. This function should modify (if necessary) 
    heap in-place.
    
    """
    #Find the indexes for the children for a given parent
    l = left(i)
    r = right(i)
    
    
    #if we haven't reached a leaf and the parent is smaller than it left children
    if l <= len(heap)-1 and heap[l] > heap[i]:
        largest = l #assign the index of the children as the largest value    
    else: 
        largest = i #otherwise, assign the index of the parten to the largest value            
    
    #if we haven't reached a leaf and the parent(or child) is smaller than it right children
    if r <= len(heap)-1 and heap[r] > heap[largest]:
        largest = r #assign the index of the children as the largest value


    #If any of the children are larger than the parent
    if largest != i:
        
        #swap the parent and the children of highest value
        heap[i], heap[largest] = heap[largest], heap[i]
        
        #call function again on the new children (previous parent)
        heapify(heap, largest)
    return
    raise NotImplementedError()
```

```python
A = [39, 85, 85, 16, 49, 7, 49, 92, 76, 15, 21, 30, 29, 31, 28]
heapify(A,0)
assert(A == [85, 49, 85, 16, 39, 7, 49, 92, 76, 15, 21, 30, 29, 31, 28])
```

## Question 3 [time estimate: 3 minutes] 

Next, write a Python implementation of the BUILD_MAX_HEAP operation using the pseudocode provided in Section 6.3 of Cormen et. al. Test your Python implementation using the array in problem 1, and make sure your Python codes produce a valid max heap.

```python
def build_max_heap(A):
    """
    Parameters
    ----------
    A : a list of floats
    
    Note
    ----
    No output is needed. The function should turn A into 
    a valid max heap, in-place.
    
    """
    #Loops through all the parents in A
    for i in range((len(A)-1)//2, -1, -1):
        
        #Calls heapify on each of the parent
        heapify(A, i)
    return
    raise NotImplementedError()
```

```python
A = [4, 1, 3, 2, 16, 9, 10, 14, 8, 7]
build_max_heap(A)
assert(A == [16, 14, 10, 8, 7, 9, 3, 2, 4, 1])
```

```python
def min_heapify(heap, i):
    """
    Parameters
    ----------
    heap : a list of floats
        Assume that the heap size is the length of the heap.
    
    Note
    ----
    No output is needed. This function should modify (if necessary) 
    heap in-place.
    
    """
    #Find the indexes for the children for a given parent
    l = left(i)
    r = right(i)
    
    
    #if we haven't reached a leaf and the parent is smaller than it left children
    if l <= len(heap)-1 and heap[l] < heap[i]:
        smallest = l #assign the index of the children as the largest value    
    else: 
        smallest = i #otherwise, assign the index of the parent to the smallest value            
    
    #if we haven't reached a leaf and the parent(or child) is smaller than it right children
    if r <= len(heap)-1 and heap[r] < heap[smallest]:
        smallest = r #assign the index of the children as the smallest value

    #If any of the children are smaller than the parent
    if smallest != i:
        
        #swap the parent and the children of smallest value
        heap[i], heap[smallest] = heap[smallest], heap[i]
        
        #call function again on the new children (previous parent)
        min_heapify(heap, smallest)
    return
    raise NotImplementedError()
```

```python
def build_min_heap(A):
    """
    Parameters
    ----------
    A : a list of floats
    
    Note
    ----
    No output is needed. The function should turn A into 
    a valid min heap, in-place.
    
    """
    #Loops through all the parents in A
    for i in range((len(A)-2)//2, -1, -1):
        
        #Calls heapify on each of the parent
        min_heapify(A, i)
    return
    raise NotImplementedError()

A = [39, 85, 85, 16, 49, 7, 49, 92, 76, 15, 21, 30, 29, 31, 28]
build_min_heap(A)
```
